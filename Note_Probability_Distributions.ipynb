## Probability Distributions

- Defines all possible values and probabilities that random variable will take.
- It maps out likelihood of multiple outcomes in a table or equation.
- A table with all possible value of a random variable & its corresponding probabilities is called as probability distribution.


- e.g. in coin flip only 2 possible outcomes H and T & each outcome has a probability of 50%. Hence a collection of these 2 outcomes along with their probability will form a probability distribution for a random experiment of a coin flip.
- e.g. if u flip 2 coins together then S is HH, TT, HT or TH & probability of each is 25%. Hence this collection of 4 outcomes along with probability will form a probability distribution.

### Measures of Distribution / Descriptive Statistic of Data Distribution

1. Skewness
2. Kurtosis

- Skewness = In probability theory & statistics skewness is measure of asymmetry of probability distribution of random variables about its mean.
- Skewness value can be +ve, -ve, 0 or in some cases undefined.

- Kurtosis = It determines the heaviness of distribution tails.
- A +ve value tells u that u have heavy tails i.e. a lot of data is in tail of distribution.
- A -ve value means u have light tails i.e. little data on the tails.
- Distribution can be divided into distinct categories based on kurtosis- 

- **Leptokurtic distribution** have +ve kurtosis >3. Tails r heavier than normal distribution.
- **Platykurtic distribution** have -ve kurtosis <3. Tails r thin compared to normal distribution or in some cases non-existent.

## 5. Exponential Distribution

- The exponential distribution is the only continuous distribution that is memoryless (or with a constant failure rate). Geometric distribution, its discrete counterpart, is the only discrete distribution that is memoryless. 
- The exponential distribution is memoryless because the past has no bearing on its future behavior. Every instance is like the beginning of a new random period, which has the same distribution regardless of how much time has already elapsed.
- The exponential distribution is often concerned with the amount of time until some specific event occurs. For example, the amount of time (beginning now) until an earthquake occurs has an exponential distribution.
- Analysts frequently use the exponential distribution to model the amount of time between independent events. In this type of study, the distribution assumes that events occur at a constant rate on average over time.
- The Poisson distribution deals with the number of occurrences in a fixed period of time, and the exponential distribution deals with the time between occurrences of successive events as time flows by continuously.
- The exponential distribution predicts the wait time until the *very first* event.



### Z-Score / Standard Score

- It gives an idea of how far from the mean a data point is in a distribution.
- It is measure of how many std dev. below or above the value is from population mean.
- z score helps u when u want to do area under the curve analysis.
- z score is typically nearer to 0 when ur predicted value & mean is v.near to one another.
