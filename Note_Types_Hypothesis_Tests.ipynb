## Types of Hypothesis Tests

### Hypotheis Tests can be -
- one sample or 2 sample or more
- one-tailed or 2-tailed
- these are generally for either mean, variance or proportion

## A/B Testing

1. A/B testing is a statistical way of comparing 2 or more versions such as version A or version B.
2. To understand not only which version performs better but also if difference b/w 2 versions is statistically significant.
3. Callout ur criteria for success ahead of time.
4. e.g. changing layout of page in 2nd version compared to 1st. 2 different images, which has a higher click through rate?
5. Ur UI/UX is same but u update ur ML algorithm to update the recommnedations that r shown to people. 

**A/B Testing Case Study Example**

Analysing 2 different website implementations & statistical significance of newer design on website traffic metrics e.g. CTR Click Through Rate.<br>
Sampling distribution is always w.r.t some statistic and in this case the statistics is the difference in proportions of click through rate for control and experimental group.

## T-test
- It is used to determine if there is significant difference b/w means(averages) of 2 groups which may be related in certain features.
- T-test is used to answer a research question.For e.g. a drug company may want to test a new cancer drug to find out if it improves life expectancy of patients.

- **CONTROL GROUP** - refers to group of people not given new drug (normal old drug used)<br>
  **NEW DRUG GROUP/TREATMENT GROUP** - refers to people tested on new drug.<br>
- It could seem from above results that drug might work but it could be due to random chance. To test this researchers can use T-test to find out if results are repeatable for entire population.
- Use Paired t-test for **before treatment** and **after treatment** for same group of people.

## ANOVA

- It is a statistical method used to test if atleast one or more of 2 groups have a statistically different mean.
- Ho (null hypothesis) = means of all groups are same.
- Ha (alternative hypothesis) = mean of atleast 1 group is different.

## Chi-Square Goodness of fit statistical hypothesis test

- In this type of hypothesis test, you determine whether the data fit a particular distribution or not.
- The term 'goodness of fit' in this test is used to compare observed sample distribution with expected probability distribution.
- To verify if our distribution of sample data is inline with some expected distribution of that data
- Ho = there is no significant difference b/w observed and expected value.
- Ha = there is a significant difference b/w observed and expected value.

### Theoritical distribution
These are distributions based on logic or mathematical formula's.

### Empirical distribution
These are frequency distributions of observed values/actual values.<br>
It is based on observations from an experiment.<br>
In real world, mostly data is normally distributed i.e. follows normal distribution.
