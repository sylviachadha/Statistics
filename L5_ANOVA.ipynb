For ANOVA, we need each group which needs to be compared, in the form of list.

```python
IS1 = [1.26,
1.31,
1.26,
1.27,
1.24,
1.3,
1.28,
1.24]
```

```python
IS2 = [1.21,
1.21,
1.24,
1.22,
1.2,
1.24]
```

```python
IS3 = [1.21,
1.19,
1.26,
1.28,
1.26,
1.24,
1.22]
```

```python
IS4 = [1.29,
1.32,
1.36,
1.39,
1.3,
1.32,
1.37]
```

```python
IS5 = [1.28,
1.35,
1.34,
1.29,
1.35]
```

```python
f_oneway(IS1,IS2,IS3,IS4,IS5)
```
