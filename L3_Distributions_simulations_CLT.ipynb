## 2. Simulating numbers from BINOMIAL DISTRIBUTION

THis distribution is defined by 2 parameters n(num of trials) and p(prob of success in each trial)

To verify properties of Binomial RV which are Expected value and Variance of RV.

```python
#binomial(n, p, size=None)
# 1 Binomial exp. itself involves doing something n times.
#Draw samples from a binomial distribution.

X_rand = np.random.binomial(100, 0.8, size = 10000)
print("One Binomial experiment itself involves doing something n times\n")
print('n=100,number of trials in each binomial experiment')
print('p=0.8, probability of success in each trial & here 100 trials in 1 binomial exp ')
print('size or num of times binomial exp repeated', len(X_rand))

print('\n10000 results of 10000 binomial exp & each result is made up of num of success in 100 trials', X_rand)
```

For Binomial Distribution-

If we calculate by Binomial RV formula, E(X) = np = 100*0.8 = 80

If we calculate by Binomial RV formula, variance = np(1-p) = 100*0.8(1-0.8) = 16


```python
# The average value of Xrand list should be close to expected value.
print(type(X_rand))
mean1 = np.mean(X_rand)
print(mean1)
```

```python
# The variance of Xrand (RV)-calculated by formula should be close to normal descriptive
# statistics variance as these numbers are simulated from binomial distribution.

var1 = np.var(X_rand)
print(var1)
```

Hence from above calculations we observe that mean comes close to 80 and variance close to 16 as expected because we simulated numbers from binomial random variable.

The distplot represents the univariate distribution of data i.e. data distribution of a variable against the density distribution.

distplot() is used to visualize the parametric distribution of a dataset.

The x-axis is number of success for each binomial experiment which inside consists of n trials.
The y-axis in a density plot is the probability density function for the kernel density estimation. For Binomial probability mass function is given as above nCr(p)^r(1-p)n-r.


```python
# Plot

sns.distplot(X_rand);
```

In plot we can see mostly around average (80) but sometimes extreme value also.

```python
ax = sns.displot(X_rand)
ax.set(xlabel="Binomial Exp outcomes-number of successes i.e. value that RV takes", 
       ylabel = "Count of each outcome");
```

## 3. Simulating numbers from POISSON DISTRIBUTION
To verify properties of Poisson RV which are Expected value and Variance of RV.

This distribution is defined by parameter lambda

```python
#poisson(lambda=1.0, size=None) # size is how many times u want to repeat experiment
#Draw samples from a Poisson distribution.

X_rand1 = np.random.poisson(50, size = 10000)   # lambda or avg value is 50 people coming
X_rand1
```

In Poisson Random Variable, Expected value and variance both equal to lambda.



```python
# Average
mean2 = np.mean(X_rand1)
print(mean2)
```

```python
# Variance
var2 = np.var(X_rand1)
print(var2)
```

```python
# Plot

sns.distplot(X_rand1);
```

### Central Limit theoram, samples drawn from different types of population distributions like Normal, Uniform, Poisson, Bernoulli and each dustribution with different number of samples.

```python
#Central Limit Theorem - Normal Population
a=[]
b=[]
c=[]
d=[]
e=[]
f=[]
for i in range(10000):   # loc refers to mean, scale refers to std dev.,
  X_rand = np.random.normal(loc = 10, scale = 5, size = 1)
  a.append(sum(X_rand)/len(X_rand))
  Y_rand = np.random.normal(loc = 10, scale = 5, size = 5)
  b.append(sum(Y_rand)/len(Y_rand))
  Z_rand = np.random.normal(loc = 10, scale = 5, size = 10)
  c.append(sum(Z_rand)/len(Z_rand))
  U_rand = np.random.normal(loc = 10, scale = 5, size = 30)
  d.append(sum(U_rand)/len(U_rand))
  V_rand = np.random.normal(loc = 10, scale = 5, size = 50)
  e.append(sum(V_rand)/len(V_rand))
  W_rand = np.random.normal(loc = 10, scale = 5, size = 100)
  f.append(sum(W_rand)/len(W_rand))

plt.figure(figsize=(10,7))
plt.hist(a, bins = 100, label = 'n=1', alpha = 0.5, density = True)
plt.hist(b, bins = 100, label = 'n=5', alpha = 0.5, density = True)
plt.hist(c, bins = 100, label = 'n=10', alpha = 0.5, density = True)
plt.hist(d, bins = 100, label = 'n=30', alpha = 0.5, density = True)
plt.hist(e, bins = 100, label = 'n=50', alpha = 0.5, density = True)
plt.hist(f, bins = 100, label = 'n=100', alpha = 0.5, density = True)
plt.title('Samples from Normal Distributions')
plt.legend()
plt.show();


# density = True means area under curve will add upto 1.
# Number of experiments = 10,000
# n=1 means each single exp out of those 10k exp is carried out with 1 sample
# n=50 means each single exp out of those 10k exp is carried out with 50 samples
```

```python
#Central Limit Theorem - Uniform Population
a=[]
b=[]
c=[]
d=[]
e=[]
f=[]
for i in range(100000):
  X_rand = np.random.uniform(low = 10, high = 50, size = 1)
  a.append(sum(X_rand)/len(X_rand))
  Y_rand = np.random.uniform(low = 10, high = 50, size = 5)
  b.append(sum(Y_rand)/len(Y_rand))
  Z_rand = np.random.uniform(low = 10, high = 50, size = 10)
  c.append(sum(Z_rand)/len(Z_rand))
  U_rand = np.random.uniform(low = 10, high = 50, size = 30)
  d.append(sum(U_rand)/len(U_rand))
  V_rand = np.random.uniform(low = 10, high = 50, size = 50)
  e.append(sum(V_rand)/len(V_rand))
  W_rand = np.random.uniform(low = 10, high = 50, size = 100)
  f.append(sum(W_rand)/len(W_rand))

plt.figure(figsize=(10,7))
plt.hist(a, bins = 50, label = 'n=1', alpha = 0.5, density = True)
plt.hist(b, bins = 50, label = 'n=5', alpha = 0.5, density = True)
plt.hist(c, bins = 50, label = 'n=10', alpha = 0.5, density = True)
plt.hist(d, bins = 50, label = 'n=30', alpha = 0.5, density = True)
plt.hist(e, bins = 50, label = 'n=50', alpha = 0.5, density = True)
plt.hist(f, bins = 50, label = 'n=100', alpha = 0.5, density = True)
plt.title('Samples from Uniform Distributions')
plt.legend()
plt.show();

# Number of experiments = 10,000
# n=1 means each single exp out of those 10k exp is carried out with 1 sample
# n=50 means each single exp out of those 10k exp is carried out with 50 samples
```

```python
#Central Limit Theorem - Poisson Population
a=[]
b=[]
c=[]
d=[]
e=[]
f=[]
for i in range(10000):
  X_rand = np.random.poisson(50, size = 1)
  a.append(sum(X_rand)/len(X_rand))
  Y_rand = np.random.poisson(50, size = 5)
  b.append(sum(Y_rand)/len(Y_rand))
  Z_rand = np.random.poisson(50, size = 10)
  c.append(sum(Z_rand)/len(Z_rand))
  U_rand = np.random.poisson(50, size = 30)
  d.append(sum(U_rand)/len(U_rand))
  V_rand = np.random.poisson(50, size = 50)
  e.append(sum(V_rand)/len(V_rand))
  W_rand = np.random.poisson(50, size = 100)
  f.append(sum(W_rand)/len(W_rand))

plt.figure(figsize=(10,7))
plt.hist(a, bins = 40, label = 'n=1', alpha = 0.5, density = True)
plt.hist(b, bins = 40, label = 'n=5', alpha = 0.5, density = True)
plt.hist(c, bins = 40, label = 'n=10', alpha = 0.5, density = True)
plt.hist(d, bins = 40, label = 'n=30', alpha = 0.5, density = True)
plt.hist(e, bins = 40, label = 'n=50', alpha = 0.5, density = True)
plt.hist(f, bins = 40, label = 'n=100', alpha = 0.5, density = True)
plt.title('Samples from Poisson Distributions')
plt.legend()
plt.show;

# Number of experiments = 10,000
# n=1 means each single exp out of those 10k exp is carried out with 1 sample
# n=50 means each single exp out of those 10k exp is carried out with 50 samples
```
