## Inferential Statistics

Like Descritive statistics helps to describe data, inferential statistics allows us to make predictions or inferences from that data.<br>
With Inferential statistics u take data in form of samples and make generalizations about population.<br>


It has two areas which are -
### 1. Estimating Parameter
It means u take some statistics from sample data (e.g. sample mean) and use it to say something about population parameter (e.g. population mean).
### 2. Hypothesis Testing
In hypothesis testing we use sample data to answer some research quetions. e.g. if a new cancer drug is effective?

## Sampling Distribution


The sampling distribution considers the distribution of sample statistics (e.g. mean or proportion), whereas the sample distribution is basically the distribution of the sample taken from the population.A sampling distribution is a statistic that is arrived out through repeated sampling from a larger population.


In general, a **sampling distribution will be normal if** either of two characteristics is true:
- the population from which the samples are drawn is normally distributed or.
- the sample size is equal to or greater than 30.

## Hypothesis Testing

1. HT is a way to test the results of an experiment to see if results r meaningful. HT determines whether results r valid by understanding the odds that the results have happened by chance. 
2. Hypothesis testing is the act of testing a hypothesis or a supposition in relation to a statistical parameter. 
3. Analysts implement hypothesis testing in order to test if a hypothesis is plausible (new knowledge is repeatable & holds true for entire population) and that these results are not due to a random chance. 
4. Hypothesis testing is a way of making statistical decisions using observed or experimental data.<br>
5. It is used for conflict resolution.



## Purpose of Hypothesis testing

HT determines whether results r valid by understanding the odds that the results have happened by chance. 

Purpose of hypothesis testing is to test whether null hypothesis can be rejected or approved.

## Key Terms in Hypothesis Testing

Every experiment will have 2 mutually exclusive hypothesis namely Ho-null hypothesis and Ha alternative hypothesis.

**Null Hypothesis (Ho)** - Status quo / existing state of affairs

**Alternative Hypothesis (Ha)** - Claim that is coming in / what u want to prove / new knowledge

### 1. Alternative Hypothesis
1. Alternative Hypothesis (H1) or the research hypothesis states that there is a relationship between two variables (where one variable affects the other). The alternative hypothesis is the main driving force for hypothesis testing. 
2. It implies that the two variables are related to each other and the relationship that exists between them is not due to chance or coincidence. 
3. When the process of hypothesis testing is carried out, the alternative hypothesis is the main subject of the testing process. The analyst intends to test the alternative hypothesis and verifies its plausibility.

**Note**<br>
New knowledge discovered when u accept alternate hypothesis or reject null hypothesis.


### 2. Null Hypothesis
1. The Null Hypothesis (H0) aims to nullify the alternative hypothesis by implying that there exists no relation between two variables in statistics. It states that the effect of one variable on the other is solely due to chance and no empirical cause lies behind it

**Note**<br>
No new knowledge discovered. Original results hold true when u fail to reject null hypothesis.


### Some Rules

1. There should be no common possibility in null and alternative hypothesis.

2. There should be no possibility outside of Ho and Ha.

3. Philosophy of testing - Innocent until **proven** guilty meaning hypothesis test is conducted under the assumption that null hypothesis is innocent/true until proved guilty/wrong.

4. Proven meaning - it depends on the criteria set by the decision maker.

## Alpha / Significance Level

- It is the probability of rejecting null hypothesis when it is true.
- It represents an acceptable probability of Type 1 error in a statistical test
- Because alpha corresponds to a probability, it can range from 0 to 1.

## Types of Errors in Hypothesis Testing

There r 2 types of errors in Hypothesis Testing.

**Type 1 error - False Positive**<br>
e.g. an innocent subjected guilty in a court of justice<br>
Null hypothesis is rejected, even though it is true and should not be rejected. 

 
**Type 2 error - False Negative**<br>
e.g. a criminal set free in a court of justice<br>
Null hypothesis is accepted, even though it is false and should be rejected.

Type 1 error rejects the Ho that is actually true in population while type 2 error fails to reject a null hypothesis that is actually false in the population.

## p-value Interpretation for Hypothesis tests

1. Whenever u see a p-value it means some statistical hypothesis test is going on in the background.
2. If p is low than ur decided significance level then null must go. So u reject the null hypothesis implying new knowledge is discovered (the claim or alternative hypothesis is accepted).
3. If p is high null fly i.e. u fail to reject Ho and existing knowledge prevails.
