### Boxplot
1. You need to sort data first in ascending order.
 
2. For eg, out of total 1000 records, the VALUE for 250th record will be 25th percentile. The VALUE for 500th record is 50th percentile or median (middle value of data),VALUE for the 750th record would represent 75th percentile.

3. 50% of data lie b/w 25th percentile and 75th percentile and this is also known as IQR - Interquartile Range. So this IQR is your box in boxplot (value of 25th percentile represents top point of box & value of 75th percentile represents bottom point of box)and the line in b/w this box is median (which would represent value of 50th percentile)

4. Interpretation of IQR - *50% of data would lie in this*. e.g from below statistics-

count         843, mean      3305054, std       4534742, min        414000, 25%        430325, 50%       1175000, 75%       4306250, max      32000000

From above data we see 50% of players are earning b/w 430325 to 4306250 - interpreation of IQR

5. Further u extend boxplot uptil availablity of data, in this case, line might go too long after the box so that's y in usual boxplot the box is extended upto a line which is 1.5 times IQR. These are called as Whiskers. Anything beyound 1)within the box (iqr) and beyond 2) whiskers (1.5 IQR) is an outlier.

```python
bp = sns.boxplot(y = rawdata.Salary, showmeans = True)  
# will show salary on y-axis,bp as per salary
```

Ques - y we cant see a line (whisker) at bottom & only see at top ? there might not be available data points. it can extend to max of 1.5IQR but it does not mean it will always, depends either 1.5IQR or availability of last data point.

Ques - y median is not in center of boxplot? Median not in center of box implies skewed data. So median value means half the observations (25%) should be above this value and half observations (25%) below this value. Now median is somewhere 0.15 here and box is appx b/w 0.1 and 0.4 meaning maximum (i.e. half players 25%) have lower salaries just upto 1.5 & rest half (another 25%) have b/w 0.15 & 4.




Green triangle is mean/average. Average can be anywhere (outside/inside box) etc depending on data point values.

```python
bp = sns.boxplot(x = rawdata.Salary, showmeans = True)  
# will show salary on x-axis,bp as per salary
```

## Multiple Boxplots
A boxplot becomes much more powerful when you view multiple boxplots together.

```python
sns.set_style('whitegrid')
bp = sns.boxplot(x = rawdata.Salary, y = rawdata.Position, showmeans = True)
bp.figure.set_figwidth(15.27)
bp.figure.set_figheight(5.7)
bp;
```

## Questions based on above boxplot

1. In which position are players making maximum salary ? Designated Hitter as green triangle represented mean is higher means average salary is more for this position compared to others. Also median is higher. It does not have outliers so risk also a bit on lower side.

2. Why is Infielder outside the box as well as outside 1.5 IQR also ? Because of that outlier, that 1 person salary is influencing so much that it takes average so up (even greater than all other player salaries). (like ambani)

3. In Infielder why is whisker small & not 1.5 times? cz that would be last available data point in 1.5 IQR range. Although we have a data point after (but that is outlier). Within 1.5IQR range that was last data point so length of whisker only upto that instead of 1.5IQR.



```python
sns.set_style('whitegrid')
bp = sns.boxplot(y = rawdata.Salary, x = rawdata.Position, showmeans = True)
bp.figure.set_figwidth(15.27)
bp.figure.set_figheight(5.7)
bp;
```

```python
sns.set_style('whitegrid')
bp = sns.boxplot(x = rawdata.Salary, y = rawdata.Pitcher, showmeans = True)
bp.figure.set_figwidth(15.27)
bp.figure.set_figheight(5.7)
bp;
```

```python
sns.set_style('whitegrid')
bp = sns.boxplot(y = rawdata.Salary, x = rawdata.Position, hue=rawdata.Yankees)
bp.figure.set_figwidth(15.27)
bp.figure.set_figheight(5.7)
bp;
```
