## Symmetric Distribution

1. In Normal Distribution Mean=Median=Mode.
2. For a normal distribution, 68% of the observations are within +/- one standard deviation of the mean, 95% are within +/- two standard deviations, and 99.7% are within +- three standard deviations.
3. For a Standard Normal Distribution mean=0 and standard deviation=1.

## Skewed Distributions


Skewness tells us about the direction of outliers. You can see that our distribution is positively skewed and most of the outliers are present on the right side of the distribution. Note: The skewness does not tell us about the number of outliers. It only tells us the direction.

## Right Skewed Distribution

1. The skew is always w.r.t tail so right skewed means the long tail is towards right. 
2. The mean is closer to the tail in a skewed distribution.
3. So in a right skewed distribution (the tail points right on the number line) i.e. moving towards more +ve direction, the mean is higher than the median.
4. Mean is to the right of median.

## Left Skewed Distribution

1. The skew is always w.r.t tail so left skewed means the long tail is towards left. 
2. The mean is closer to the tail in a skewed distribution.
3. So in a left skewed distribution (the tail points left on the number line) i.e. moving towards more -ve direction, the mean is lower than the median.
4. Mean is to the left of the median
