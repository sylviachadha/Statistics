**Confidence Interval**<br>
Confidence intervals show us the likely range of values of our population mean. When we calculate the mean we just have one estimate of our metric; CI give us richer data and show the likely values of the true population mean.

Since n is large (>30),as per CLT,irrespective of original population distribution, sample mean distribution will appx. a normal distribution.


```python
# Sample size, mean, standard deviation, standard error.
n = len(df2)
xbar = df2['Annual Income'].mean()    
s = df2['Annual Income'].std(ddof=1)# for sample std
std_error = s/np.sqrt(n)

print('Number of samples=',n)
print('Mean =',xbar)
print('Std.Dev =',s)
print('Std_error =',std_error)
```

```python
s = df2['Annual Income'].std(ddof=1)
s
206649.78162315575
```


We do not make any assumption about population distribution and n>30(large), sigma is unknown so we estimate with s using below formula to build approximate CI.

**90% Confidence Interval**

```python
l_ci = round((xbar-(1.645*(s/np.sqrt(n)))),2)
u_ci = round((xbar+(1.645*(s/np.sqrt(n)))),2)

print('90% Confidence Interval-',l_ci,u_ci)
```

**95% Confidence Interval**

```python
l_ci = round((xbar-(1.96*(s/np.sqrt(n)))),2)
u_ci = round((xbar+(1.96*(s/np.sqrt(n)))),2)

print('95% Confidence Interval-',l_ci,u_ci)
```

**99% Confidence Interval**

```python
from math import sqrt

l_ci = round((xbar-(2.576*(s/np.sqrt(n)))),2)
u_ci = round((xbar+(2.576*(s/np.sqrt(n)))),2)

print('99% Confidence Interval-',l_ci,u_ci)
```

**Hypothesis Test for Proportions**

**For part 4a, One-tail test**

Ha: pamd > panand<br>
Ho: pamd <= panand


**Calculate using python function-proportions_ztest**

This returns z-statistic and p-value as result.

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np

sig1 = 0.01
sig2 = 0.05
sig3 = 0.10


# note - the samples do not need to be the same size
sample_success_a, sample_size_a = (60, 300)
sample_success_b, sample_size_b = (90, 400)


# check our sample against Ho to check for Ha > Ho
successes = np.array([sample_success_a, sample_success_b])
samples = np.array([sample_size_a, sample_size_b])


# note, no need for a Ho value here(as for 1 sample) - it's derived from the other parameters.
# for Ha > Ho use alternative='larger'
# for Ha < Ho use alternative='smaller'
# for Ha != Ho use alternative='two-sided'
stat, p_value = proportions_ztest(count=successes, nobs=samples,  alternative='larger')

# Report result
print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))
if p_value > sig1:
   print ("Fail to reject the null hypothesis at 99% Confidence level")
else:
   print ("Reject the null hypothesis at 99% Confidence level")
if p_value > sig2:
   print ("Fail to reject the null hypothesis at 95% Confidence level")
else:
   print ("Reject the null hypothesis at 95% Confidence level")
if p_value > sig3:
   print ("Fail to reject the null hypothesis at 90% Confidence level")
else:
   print ("Reject the null hypothesis at 90% Confidence level")
```

**For part 4b, two-tail test**

Change alternative='two-sided'here. In 4a we use alternative='larger'.<br>
Ho: p_amd = p_anand  or  p_amd-p_anand=0 <br>
Ha: p_amd != p_anand

```python
# note, alternative='two-sided' for 2 tail test
stat1, p_value1 = proportions_ztest(count=successes, nobs=samples,  alternative='two-sided')


# Report result
print('z_stat: %0.3f, p_value: %0.3f' % (stat1, p_value1))
if p_value1 > sig1:
   print ("Fail to reject the null hypothesis at 99% Confidence level")
else:
   print ("Reject the null hypothesis at 99% Confidence level")
if p_value1 > sig2:
   print ("Fail to reject the null hypothesis at 95% Confidence level")
else:
   print ("Reject the null hypothesis at 95% Confidence level")
if p_value1 > sig3:
   print ("Fail to reject the null hypothesis at 90% Confidence level")
else:
   print ("Reject the null hypothesis at 90% Confidence level")
```
