### Building blocks of Probability

### 1. Random Experiment

- Probability is based on random experiments. And Random Experiments r one which r performed under identical conditions & whose results cannot be predicted.
- In Probability theory an experiment or trial is any procedure that can be infinitely repeated & has a well defined set of possible outcomes known as sample space.

Egs of Random Experiments-
- Tossing coin
- Throwing dice
- Drawing cards

### Difference b/w Deterministic and Non-deterministic/random experiments

- A deterministic experiment is one whose outcome may be predicted with certainty beforehand. e.g. adding two numbers such as 2 3

- A Non deterministic experiment or a random experiment is an experiment for which more than one possible outcomes. Due to uncertainty, outcomes of this experiment does not predicted before performing the experiment.

### 2. Sample Space, denoted by S

It is set of all possible outcomes of a random experiment.Sample space is defined based on how u define ur random experiment.

In case of coin, S={H,T}
In case of dice, S={1,2,3,4,5,6}


### 3. Event

Any subset of a sample space is an event. 

In tossing a coin, Heads & Tails are events.

**Impossible event** {} null set<br>
**Sure Event** complete set S which is made up of all the events is called sure event

### 4. Mutually Exclusive Events

Those events that cannot take place simultaneously. When u toss a coin one time you will get either a head or a tail. hence these are mutually exclusive, cannot occur at same time.


### 5. Exhaustive Events
Those events, the union of which, form the complete set. E.g. union of H with T will form complete set.So these r exhaustive events i.e. when u r considering all the outcomes of a random experiment.
