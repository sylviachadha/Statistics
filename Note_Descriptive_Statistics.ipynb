## Descriptive Statistics

Descriptive statistics describes the data. <br>
4 parts to learning statistical thinking in Data Science


Descriptive Statistics-

- describes basic features of data
- provides summary about sample and measures
- forms basis of quantitative analysis

## 1. Measures of Central Tendency - Mean, Median and Mode

Measures of Central tendency is a single value that attempts to describes a set of data by identifying central position within that set of data. Central tendency can be calculated using 3 distinct measures - Mean, Median, Mode.




1. **Mean** = describes an entire sample with a simple score that represents the arithmetic center of data/arithmetic avg of data.
2. **Median** = divides data into exactly similar parts i.e. it gives the value located at center of data.
3. **Mode** = value that occurs most frequently.

**Case where median preferred over Mean**

Income distribution in country where 20% rich people roughly equal income of 80% remaining people. In this case mean will not give true picture as it will get jacked up due to high income of 20% people. Median will be more realistic (arrange income in ascending order & choose the middle value).

** In retail, mode is helpful to understand customer choice**

## 2. Measures of Spread/Dispersion - Range, Mean absolute deviation, Variance, Standard Deviation.



- Measures of dispersion indicate the scattering of data.
- gives idea about variation in data.
- Delivering a view on distribution of data.

1.**Range** = Maximum value in data-Minimum value in data.<br>
2.**Mean Deviation** = mean or avg of deviations of observations from central value. The central value can be either mean or median.<br>
3. **Standard Deviation** = square root of arithmetic avg of squares of deviations measured from mean.<br>
4. **Variance** = Squared value of std deviation.

### 3. Measures of Position

- Measure of Position help us partition the data.
- Percentile can be used in ranking, if percentile is 50th percentile it means that it performs better than 50% of entities within its group.
- Quartiles can be used in survey data to divide population into groups & calclate tax they will have to pay according to quartile they fall in.



### 4. Correlation Statistical technique

1. It is a statistical technique that can show whether and how strongly pairs of variables are related with each other.
2. More closer to each other forming straight line means strongly correlated while more scattered they become then weakly correlated. If totally random pattern then no correlation.
