## 2. Chi-Square Distribution

#### Observations

1. As seen above Chi-Square Random variable U, is sum of squares of std normal random variables (n degree of freedom).

2. Also below plot verifies properties of chi-square distribution, we can see mean to be around 4 (= degrees of freedom) and varince (2*4=8) so plot extending about till 12 appx.

3. Degree of freedom=4 means u have 4 random variables which are free to vary as they like.

```python
#chi-square with n-degrees of freedom here n=4 so 4 degrees of freedom

n = 4   # means 4 standard normal random variables
Y= []
for i in range(100000):
    U_chi = 0
    for j in range(n):
        Z = np.random.normal(0,1) # Z1, Z2, Z3, Z4
        Z = Z**2     # Z1^2, Z2^2, Z3^2, Z4^2
        U_chi = U_chi + Z   # As per definition of chi-square sum of squares of k independent 
    Y.append(U_chi)       # std normal rv.

p=sns.histplot(Y)
p.set_xlabel('Chi-square RV values')
p.set_title('Chi-Square Distribution Simulation')
plt.show()
print(len(Y))
```

## 3. t-Distribution

As n starts getting larger & larger, t distribution starts getting approximated as normal distribution.
So if degrees of freedom r vv large this is almost same as normal distribution.

As seen in below figure also, <br>
mean=0 <br>
Variance = v/v-2 => 10/8 => sd = 1.1 => 3 sd = 3.3

```python
#t-distribution with n-degrees of freedom
n = 10
Y= []
for i in range(100000):
    chi = 0
    for j in range(n):
        Z = np.random.normal(0,1)
        Z = Z**2
        chi = chi + Z
    t = np.random.normal(0,1)   
    t = t/((chi/n)**0.5)
    Y.append(t)    

sns.histplot(Y)
plt.show()
```

## 4. F Distribution

```python
#F-distribution with m and n-degrees of freedom
m=100
n = 100
Y= []
for i in range(100000):
    chi1 = 0
    chi2 =0
    for j in range(m):
        Z = np.random.normal(0,1)
        Z = Z**2
        chi1 = chi1 + Z
    for j in range(n):
        Z = np.random.normal(0,1)
        Z = Z**2
        chi2 = chi2 + Z    
    f = (chi1*n)/(chi2*m)
    Y.append(f)    

sns.histplot(Y)
plt.show()
```
