### If 3% is positivity rate, calculate appropriate number of samples to be pooled?

```python
expectedtests = {}
for k in range(2,500):    # put some sensible number(2,500), 1000 tests if individually
  expectedtests[k]= (1000/k + 1000*(1-(0.97**k)))  # updating dict for different values of k
```

#### Find key corresponding to a specific value in dictionary

```python
# find(key - which is no of samples) corresponding to this value is -

keys = [key for key, value in expectedtests.items() if value == 333.6946617376668]
print(keys)
```

Hence if 3% is positivity rate, appropriate number of samples to be pooled together is 6.
